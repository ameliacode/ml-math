```python
import numpy as np
```

```python
input_data = np.array([[2, 3], [5, 1]])
x = input_data.reshape(-1)
```

```python
w1 = np.array([2, 1, -3, 3])
w2 = np.array([1, -3, 1, 3])
b1 = 3
b2 = 3
```

```python
W = np.array([w1, w2])
W
```

```python
b = np.array([b1, b2])
b
```

```python
weight_sum = np.dot(W, x) + b
weight_sum
```

```python
res = 1 / (1 + np.exp(-weight_sum))
res
```
